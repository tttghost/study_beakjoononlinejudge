```python
# 아이디어 도출단계,  bfs를 바로 떠올리지 못했다. 가장 짧은거리는 bfs, queue넣고 돌고 반복!
from collections import deque

EMPTY='.'
WALL='#'
PLAYER='@'
FIRE='*'
FAIL='IMPOSSIBLE'

N=int(input())

# q=deque()

#플레이어를 찾는다.

dx=[-1,1,0,0]
dy=[0,0,-1,1]


for i in range(1):
    col,row=map(int,input().split())
    building=[input() for _ in range(row)]
    player_visited=[list(True if col==PLAYER else False for col in row) for row in building]
    fire_visited=[list(True if col==FIRE else False for col in row) for row in building]
    success=False
    while not success:
        for y in range(row):
            if success:
                break
            for x in range(col):
                servive=False
                if building[y][x] == PLAYER:
                    for k in range(4):
                        nx,ny=x+dx[k],y+dy[k]
                        if building[ny][nx]==EMPTY:
                            if nx==0 or nx==col-1 or ny==0 or ny==row-1:
                                success=True
                                break
                            else:
                                servive=True
                                building[ny][nx]=PLAYER
                        
                        
                if building[y][x] == FIRE:
                    for k in range(4):
                        nx,ny=x+dx[k],y+ny[k]
                        if building[ny][nx]==EMPTY or building[ny][nx]==PLAYER:
                            building[ny][nx]=FIRE
                
                if not servive:
                    success=True
                    print(FAIL)
                    break
                    
                
```

```python
# 최종 결과물, 소요시간:5시간, 결과: 만족

initTestCase()

from collections import deque


EMPTY='.'
WALL='#'
PLAYER='@'
FIRE='*'
FAIL='IMPOSSIBLE'

dx=[-1,1,0,0] ##좌우
dy=[0,0,-1,1] ##상하

fire_q=deque()
player_q=deque()

N=int(input())

for _ in range(N):
    col,row=map(int,input().split())
    building=[list(input().rstrip()) for _ in range(row)]
    fire_q.clear()
    player_q.clear()

    for y in range(row):
        for x in range(col):
            if building[y][x]==FIRE:
                fire_q.append((y,x))
            elif building[y][x]==PLAYER:
                player_q.append((y,x))
    
    state=0 # 0 무한루프, 1성공, 2실패
    
    cnt=0

    # while state==0:
    while player_q: #이게 핵심!!! 플레이어큐가 없다?->갈수있는 환경이 아니다 -> 조기 실패
        # state=2 #위에 주석 때문에 얘는 안해줘도 됨
        cnt+=1
        
        playerLen=len(player_q)
        for _ in range(playerLen):
            if state==1:
                break
            y,x=player_q.popleft()
            if building[y][x]==PLAYER:
                for i in range(4):
                    ny,nx=y+dy[i],x+dx[i]
                    if ny<0 or ny>row-1 or nx<0 or nx>col-1:
                        state=1
                        break
                    if building[ny][nx]==EMPTY:
                        # state=0 #플레이어가 한번이라도 이동 한다면 무한루프 상태로 변환 #역시나 상태전환 필요 없음
                        building[ny][nx]=PLAYER
                        player_q.append((ny,nx))
        
        
        if state==1:
            print(cnt)
            break
        # elif state==2: #이부분을 while 밖으로 뺌
        #     print(FAIL)
        #     break
        
        fireLen=len(fire_q)
        for _ in range(fireLen):
            y,x=fire_q.popleft()
            for i in range(4):
                ny,nx=y+dy[i],x+dx[i]
                if ny<0 or ny>row-1 or nx<0 or nx>col-1:
                    continue
                if building[ny][nx]==EMPTY or building[ny][nx]==PLAYER:
                    building[ny][nx]=FIRE
                    fire_q.append((ny,nx))
    else:
        print(FAIL)
```
