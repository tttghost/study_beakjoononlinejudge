```python
import sys

operator_symbol = ['+', '-', '*', '/']

#region input
# N=int(input())
# li=[*map(int, input().split())]
# operator=[*map(int, input().split())]
#endregion

#region openfile
file=open('test.txt')
N=int(file.readline())
li=[*map(int, file.readline().rstrip().split())]
operator=[*map(int, file.readline().rstrip().split())]
file.close()
#endregion

result=''
for i,(cnt,symbol) in enumerate(zip(operator,operator_symbol)):
    result+=symbol*cnt
operSize=len(result)
chk=[False]*operSize
opers=[]

minVal=sys.maxsize
maxVal=-sys.maxsize

def dfs(cnt):

    global minVal
    global maxVal

    if cnt==operSize:
        val=li[0]
        for i,o in enumerate(opers):
            if o=='+':
                val+=li[i+1]
            elif o=='-':
                val-=li[i+1]
            elif o=='*':
                val*=li[i+1]
            elif o=='/':
                if val<0:
                    val=-(abs(val)//li[i+1])
                else:
                    val//=li[i+1]

        minVal=min(val,minVal)
        maxVal=max(val,maxVal)

        return

    for i in range(operSize):
        if chk[i]==False:
            chk[i]=True
            opers.append(result[i])
            dfs(cnt+1)
            opers.pop()
            chk[i]=False

dfs(0)
print(f'{maxVal}\n{minVal}')
```

```python
#좀더깔끔한것 보고 변형, 파라미터 2개 씀씀
# import sys
# input=sys.stdin.readline

N=int(input())
li=[*map(int, input().rstrip().split())]
operator=[*map(int, input().rstrip().split())]

operSize=sum(operator)

minVal=sys.maxsize
maxVal=-sys.maxsize

def dfs(cnt,val):

    global minVal
    global maxVal

    if cnt==operSize:
        minVal=min(val,minVal)
        maxVal=max(val,maxVal)
        return
    
    nextCnt=cnt+1
    nextVal=li[nextCnt]
    
    if operator[0]>0:
        operator[0]-=1
        dfs(nextCnt,val+nextVal)
        operator[0]+=1
    if operator[1]>0:
        operator[1]-=1
        dfs(nextCnt,val-nextVal)
        operator[1]+=1
    if operator[2]>0:
        operator[2]-=1
        dfs(nextCnt,val*nextVal)
        operator[2]+=1
    if operator[3]>0:
        operator[3]-=1
        dfs(nextCnt,int(val/nextVal))
        operator[3]+=1

dfs(0,li[0])
print(f'{maxVal}\n{minVal}')
```
