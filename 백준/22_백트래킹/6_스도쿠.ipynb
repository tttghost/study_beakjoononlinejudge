```python
#너무 심플하게 생각했음, 계속 반복하는 이상한 문제... 그리고 1개만 비어있다고 생각했음..
with open('6_input.txt') as file:
    input=file.readline
    board=[i.split() for i in file.readlines()]
full=set(str(i) for i in range(1,10))    

N=9

def dfs():
    #가로 찾고
    for i in range(N):
        if board[i].count('0') == 1:
            getdata=(full-set(board[i])).pop()
            board[i][board[i].index('0')]=getdata
            # print('a',getdata)
            dfs()
            return

    #세로 찾고
    for i in range(N):
        temp=[]
        for j in range(N):
            temp.append(board[j][i])
        if temp.count('0') == 1:
            getdata=(full-set(temp)).pop()
            board[temp.index('0')][i]=getdata
            # print('b',getdata)
            dfs()
            return

    #네모 찾고
    for i in range(N//3):
        for j in range(N//3):
            temp=[]
            for k in range(3*j,3*(j+1)):
                temp.extend(board[k][3*i:3*(i+1)])
            if temp.count('0') == 1:
                tempj=temp.index('0')%3
                tempi=temp.index('0')//3
                getdata=(full-set(temp)).pop()
                board[3*j+tempj][3*i+tempi]=getdata
                # print('c',getdata)
                dfs()
                return
                

print('\n'.join(' '.join(map(str,row)) for row in board))
print('=================')
dfs()
print('\n'.join(' '.join(map(str,row)) for row in board))
```
