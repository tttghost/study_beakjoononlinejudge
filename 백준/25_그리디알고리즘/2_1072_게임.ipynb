```python
from decimal import Decimal
from math import floor
import math




def binary_search(X,Y,pivot):
    lo,hi=1,1_000_000_000
    rs=0
    while lo<=hi:
        mid=(lo+hi)//2
        # temp=floor(Decimal(Y+mid)//Decimal(X+mid)*100)
        temp=((Y+mid)*100)//((X+mid))
        if pivot<temp:
            rs=mid
            hi=mid-1
        else:
            lo=mid+1
    return rs

        
for i in range(8):
    initTestCase(i)
    X,Y=map(int,input().split())
    # pivot=floor(float((Decimal(Y)//Decimal(X)))*100)
    pivot=(Y*100)//(X) # 소숫점 버릴 거라면 이방법이 부동소숫점 에러를 방지하는 더 효율적인 방법
    #승률이 99%이상이면 변할 수 없음
    if pivot>=99:
        print(-1)
    else:
        print(binary_search(X,Y,pivot))
```
