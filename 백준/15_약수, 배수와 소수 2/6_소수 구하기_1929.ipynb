```python
def primeList(startidx,n):
    prime=[True]*(n+1)
    prime[0:startidx]=[False]*startidx
    prime[0:2]=[False]*2
    
    for i in range(2, int(n**0.5)+1):
        for j in range(i*i, n+1, i):
            prime[j]=False
    return [i for i ,v in enumerate(prime) if v]
```

```python
M,N=map(int,input().split())
```

```python
print('\n'.join(map(str,primeList(min(M,N),max(M,N)))))
```
