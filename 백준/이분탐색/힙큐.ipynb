```python
#다익스트라 알고리즘 1일 1다익스트라 연습연습!!
initTestCase()

#노드(V)개수, 선(E)개수
V,E=map(int,input().split())

#시작지점
start=int(input())

#그래프 정보
graph=[[] for _ in range(V+1)]
for _ in range(E):
    a,b,c=map(int,input().split())
    graph[a].append([b,c])

INF=1e9
visited=[False]*(V+1) #방문여부
dist=[INF]*(V+1) #정점까지의 최소거리 계산


def get_smallest_node():
    minv=INF
    idx=0
    for i in range(1,V+1):
        if not visited[i] and minv>dist[i]:
            minv=dist[i]
            idx=i
    return idx

def dijkstra(start):
    
    #최초지점 셋팅
    dist[start]=0
    visited[start]=True
    
    #시작노드와 인접한 노드 계산
    for node in graph[start]:
        dist[node[0]]=node[1]


    #인접노드 계산         
    for i in range(1,V+1):
        #계산이 안된 가장 짧은 노드 가져오기
        now=get_smallest_node()
        nowCost=dist[now]

        #해당 노드 방문처리
        visited[now]=True

        #가장 짧은 노드와 인접한 노드 코스트 계산
        for node in graph[now]:
            nextNode=node[0]
            nextCost=node[1]
            cost=nowCost+nextCost
            if cost<dist[nextNode]:
                dist[nextNode]=cost

dijkstra(start)

for i in range(1,V+1):
    if dist[i]==INF:
        print("도달할수없음")
    else:
        print(dist[i])
```
