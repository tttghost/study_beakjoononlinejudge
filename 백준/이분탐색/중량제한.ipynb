```python
#이분탐색으로 해보기 - 이걸로 통과
#파라메트릭 서치
#bfs + binary_search

from collections import deque

initTestCase(1)

N,M=map(int,input().split())
minv=1
maxv=1000000000
graph=[[] for _ in range(N+1)]
for _ in range(M):
    a,b,c=map(int,input().split())
    graph[a].append((b,c))
    graph[b].append((a,c))

st,en=map(int,input().split())    

def bfs(mid_weight):
    dq=deque([st])
    visited=[False]*(N+1)
    visited[st]=True

    while dq:
        now_node=dq.popleft()
        if now_node == en:
            return True
        for next_node,next_cost in graph[now_node]:
            if not visited[next_node] and mid_weight<=next_cost:
                visited[next_node]=True
                dq.append(next_node)

    return False
    
result=minv

while minv<=maxv:
    mid = (minv+maxv)//2
    if bfs(mid):
        result=mid
        minv=mid+1
    else:
        maxv=mid-1

print(result)



```
