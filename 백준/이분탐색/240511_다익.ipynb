```python
#1다익 - 기본형
initTestCase()

n,m=map(int,input().split())
st=int(input())

graph=[[] for _ in range(n+1)]

visited=[False]*(n+1)
distance=[1e9]*(n+1)

for _ in range(m):
    a,b,c=map(int,input().split())
    graph[a].append((b,c))
visited[st]=True
distance[st]=0
for i in graph[st]:
    distance[i[0]]=i[1]

def get_smallest_distance():
    minv=1e9
    idx=0
    for i in range(1,n+1):
        if not visited[i] and minv>distance[i]:
            minv=distance[i]
            idx=i
    return idx

for _ in range(n):
    now=get_smallest_distance()
    visited[now]=True
    for node in graph[now]:
        cost=distance[now]+node[1]
        if cost<distance[node[0]]:
            distance[node[0]]=cost

for i in range(1,n+1):
    if distance[i]==1e9:
        print('INF')
    else:
        print(distance[i])
```

```python
#1다익 - 힙형
initTestCase()

from heapq import heappush,heappop

n,m=map(int,input().split())
st=int(input())

graph=[[] for _ in range(n+1)]
distance=[1e9]*(n+1)
distance[st]=0

for _ in range(m):
    a,b,c=map(int,input().split())
    graph[a].append((b,c))

hq=[]
heappush(hq,(0,st)) # 거리, node 순서

while hq:
    print(hq)
    dist,now = heappop(hq)
    print('꺼냄',dist,now)
    if distance[now]<dist: # 현재 노드가 이미 처리된 적이 있는 노드라면 무시
            print('무시',now,distance[now],dist)
            continue
    for node in graph[now]:
        cost=dist+node[1] #이부분 까먹지 말자!!!        
        if cost<distance[node[0]]: # 기본적으로 맨처음엔 INF임, 그렇기 때문에 갱신할 수 있음
            distance[node[0]]=cost
            print('추가',f'{dist}+{node[1]}={cost}',node[0])
            heappush(hq,(cost,node[0]))

for i in range(1,n+1):
    if distance[i]==1e9:
        print('INF')
    else:
        print(distance[i])                
```
