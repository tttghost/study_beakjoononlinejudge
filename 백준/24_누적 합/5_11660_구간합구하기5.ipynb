```python
#한방에 성공
idx=0
def input():
    global idx
    line=file[idx]
    idx+=1
    return line

file='''
4 3
1 2 3 4
2 3 4 5
3 4 5 6
4 5 6 7
2 2 3 4
3 4 3 4
1 1 4 4

'''.strip().split('\n')

output='''
27
6
64
'''
# import sys
# input=sys.stdin.readline

N,M=map(int,input().split())

A=[[0]+list(map(int,input().split())) for _ in range(N)]
A.insert(0,[0]*(N+1)) #계산을 위해 0행&0열 0 깔아줌

B=[list(map(int,input().split())) for _ in range(M)]

#점화식
for j in range(1,N+1):
    for i in range(1,N+1):
        A[j][i]+=A[j-1][i]+A[j][i-1]-A[j-1][i-1]
#결과값
for i in range(M):
    x1=B[i][0]
    y1=B[i][1]
    x2=B[i][2]
    y2=B[i][3]

    print(A[x2][y2]-A[x2][y1-1]-A[x1-1][y2]+A[x1-1][y1-1])
```
