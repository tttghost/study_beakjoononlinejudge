```python
from itertools import product #중복순열
from itertools import combinations #조합

#승 용    
def binary_search승(arr,target,low,high):
    if low>high:
        return low
    mid=(low+high)//2
    
    if arr[mid] <= target:
        return binary_search승(arr, target, mid+1, high)
    else:
        return binary_search승(arr, target, low, mid-1)
    
#패 용
def binary_search패(arr,target,low,high):
    if low>high:
        return low
    mid=(low+high)//2
    
    if arr[mid] < target:
        return binary_search패(arr, target, mid+1, high)
    else:
        return binary_search패(arr, target, low, mid-1)
    
def solution(dice):
    
    answer승 = []
    answer패 = []
    diceLen=len(dice)
    minv=-1e9
    maxv=-1e9

    role=list(combinations(range(diceLen),diceLen//2))

    for v in range(len(role)//2):
        a=role[v]
        b=role[-v-1]
        
        a_li=[]
        b_li=[]
        
        for i in a:
            a_li.append(dice[i])
        for i in b:
            b_li.append(dice[i])

        a_rs=[]
        b_rs=[]

        prod_li=list(product(*(range(6) for _ in range(diceLen//2))))

        for i in prod_li:
            temp=list(i)
            toto=0
            for j in range(len(temp)):
                toto+=a_li[j][temp[j]]
            a_rs.append(toto)
        a_rs.sort()
        
        for i in prod_li:
            temp=list(i)
            toto=0
            for j in range(len(temp)):
                toto+=b_li[j][temp[j]]
            b_rs.append(toto)
        b_rs.sort()

        total승=0
        total패=0
        for i in range(len(a_rs)):
            total승+=len(b_rs) - binary_search승(a_rs,b_rs[i],0,len(a_rs)-1)
            total패+=binary_search패(a_rs,b_rs[i],0,len(a_rs)-1)

            
        if maxv<total승:
            answer승=list(i+1 for i in a)
            maxv=total승

        if minv<total패:
            answer패=list(i+1 for i in b)
            minv=total패

    if maxv>minv:
        return answer승
    else:
        return answer패


# dice=	[[1, 2, 3, 4, 5, 6], [3, 3, 3, 3, 4, 4], [1, 3, 3, 4, 4, 4], [1, 1, 4, 4, 5, 5]] #[1,4]
dice=[    [40, 41, 42, 43, 44, 45],    [43, 43, 42, 42, 41, 41],    [1, 1, 80, 80, 80, 80],    [70, 70, 1, 1, 70, 70],    ] #예제 [[1,3]
# dice=	[[1, 2, 3, 4, 5, 6], [2, 2, 4, 4, 6, 6]] # [2]
result=solution(dice)
print(result)
        
```
